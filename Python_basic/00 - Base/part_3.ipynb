```

```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год`

````python


year = 1900

if year % 100 == 0:
    if year % 400 == 0:
        print(f"Високосный")
    else:
        print(f"Обычный")
elif year % 4 == 0:
    print(f"Високосный")
else:
    print(f"Обычный")


````

## Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Несчастливый билет`  
  
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
number = 123456
numb = str(number)


if  numb[0] + numb[1] + numb[2] == numb[3] + numb[4] +numb[5]:
    print(f"Счастливый билет")
else:
    print(f"Несчастливый билет")
````

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:  
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`  

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
import math

base = input(f"Введите фигуру (Круг, Треугольник, Прямоугольник)").lower()

if base == f"круг":
    r = int(input(f"Введите радиус круга"))
    print(f"S = {round(math.pi * r, 3)}")
elif base == f"треугольник":
    a = int(input(f"Введите длину А"))
    b = int(input(f"Введите длину В"))
    c = int(input(f"Введите длину С"))
    p = (a + b + c)/ 2
    print(f"S = {(p * (p - a) * (p - b) * (p - c)) ** 0.5}")
elif base == f"прямоугольник":
    a = int(input(f"Введите длину А"))
    b = int(input(f"Введите длину В"))
    print(f"S = {a * b}")
else:
    print(f"Непраильные данные поданы на вход")






````
